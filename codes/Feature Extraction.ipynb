# 1. Using Index Fund Nifty's Previous Day & Moving Average price as a feature

```python
df_Nifty = pd.read_csv("NIFTY.csv")
df_Nifty.set_index('Date',inplace = True)
Nifty = df_Nifty.drop(columns=['Open', 'High', 'Low', 'Volume']);Nifty
```

```python
p = df_tcs.index.difference(df_Nifty.index)
p
```

```python
Nifty.reset_index(inplace = True)
temp_df = pd.DataFrame({'Date':p, 'Close':0})
Nifty = pd.concat([Nifty, temp_df])
Nifty.sort_values(by = ['Date'], inplace = True)
#Nifty['Date'] = pd.to_datetime(Nifty['Date'])
Nifty.set_index('Date', inplace = True)
Nifty
```

```python
def replace_mean(x):
    for i in range(len(x)):
        if x[i] == 0:
            x[i] = np.mean([x[i-2],x[i-1],x[i+1],x[i+2]])
        else:
            pass
    return x
```

```python
Nifty['Close'] = replace_mean(Nifty.Close.values)
```

## Previous day Prices

```python
Nifty ['Nifty (t-1)'] = Nifty.Close.shift(periods=1)
Nifty ['Nifty (t-2)'] =  Nifty.Close.shift(periods=2)
Nifty ['Nifty (t-5)'] =  Nifty.Close.shift(periods=5)
Nifty 
```

## Moving Average Price

```python
Nifty['Nifty_MA10'] = Nifty.Close.rolling(window=10).mean()
Nifty['Nifty_MA20'] = Nifty.Close.rolling(window=20).mean()
Nifty['Nifty_MA50'] = Nifty.Close.rolling(window=50).mean()
Nifty.rename(columns = {'Close':'Nifty_Close'}, inplace = True)
Nifty
```


# 2. Bollinger Bands

### Bollinger Bands allow traders to monitor and take advantage of shifts in price volatilities

### Main Components of a Bollinger Bands


- Upper Band: The upper band is simply two standard deviations above the moving average of a stockâ€™s price.
- Middle Band: The middle band is simply the moving average of the stockâ€™s price.
- Lower Band: Two standard deviations below the moving average is the lower band.

```python
def Bollinger_Bands(data,name):
    data['MA_20'] = data.Close.rolling(window=20).mean()
    data['SD20'] = data.Close.rolling(window=20).std()
    data['Upper_Band'] = data.Close.rolling(window=20).mean() + (data['SD20']*2)
    data['Lower_Band'] = data.Close.rolling(window=20).mean() - (data['SD20']*2)
    print(data.tail())
    data[['Close', 'MA_20', 'Upper_Band', 'Lower_Band']].plot(figsize=(12,6))
    plt.title(f'20 Day Bollinger Band for {name}')
    plt.ylabel('Price (USD)')
    plt.show();
```

```python
Bollinger_Bands(bhp,'Bharat Petroleum')
```

```python
Bollinger_Bands(cipla,'Cipla')
```

```python
Bollinger_Bands(coromandel,'Coromandel')
```

```python
Bollinger_Bands(tcs,'TCS')
```

# 3. Shifting for Lagged data - Adding Previous Day prices

```python
def adding_prev_price(data,name):
    data[f'{name}_Close(t-1)'] = data.Close.shift(periods=1)
    data[f'{name}_Close(t-2)'] = data.Close.shift(periods=2)
    data[f'{name}_Close(t-5)'] = data.Close.shift(periods=5)
    data[f'{name}_Close(t-10)'] = data.Close.shift(periods=10)
    data[f'{name}_Open(t-1)'] = data.Open.shift(periods=1)
    return data.head(20)
```

# 5.Moving Average Convergance Divergance

```python
def MACD(data):
    data['EMA_12'] = data.Close.ewm(span=12, adjust=False).mean()
    data['EMA_26'] = data.Close.ewm(span=26, adjust=False).mean()
    data['MACD'] = data['EMA_12'] - data['EMA_26']
    data['MACD_EMA'] = data.MACD.ewm(span=9, adjust=False).mean()
    #print(data.tail())
    data[['MACD', 'MACD_EMA']].plot()
    plt.show()
```

# 6.Exponential Moving Averages

```python
def EMA(data):
    data['EMA10'] = data.Close.ewm(span=5, adjust=False).mean().fillna(0)
    data['EMA20'] = data.Close.ewm(span=5, adjust=False).mean().fillna(0)
    data['EMA50'] = data.Close.ewm(span=5, adjust=False).mean().fillna(0)
    data['EMA100'] = data.Close.ewm(span=5, adjust=False).mean().fillna(0)
    data['EMA200'] = data.Close.ewm(span=5, adjust=False).mean().fillna(0)
    return data.tail(10)
```

# 9.Commodity Channel Index

```python
def CCI(data):
    tp = (data['High'] + data['Low'] + data['Close']) /3
    ma = tp/20 
    md = (tp-ma)/20
    data['CCI'] = (tp-ma)/(0.015 * md)
    return data.tail(10)
```

# 13.Stochastic %K


```python
def SK(data):
    data['SO%K'] = ((data.Close - data.Low.rolling(window=14).min()) / (data.High.rolling(window=14).max() - data.Low.rolling(window=14).min())) * 100
    return data.tail(10)
```

# 18. Extract Features from Date

```python
def convert_date(data):
    data['Date_Col'] = data.index
    data['Date_Col'] = pd.to_datetime(data['Date_Col'])
    return data.shape
```

```python
print(convert_date(hdfc))
print(convert_date(bhp))
print(convert_date(cipla))
print(convert_date(coromandel))
print(convert_date(tcs))
```

```python
from datetime import datetime

def extract_date_features(date_val):
    
    Day = date_val.day
    DayofWeek = date_val.dayofweek
    Dayofyear = date_val.dayofyear
    Week = date_val.week
    Is_month_end = date_val.is_month_end.real
    Is_month_start = date_val.is_month_start.real
    Is_quarter_end = date_val.is_quarter_end.real
    Is_quarter_start = date_val.is_quarter_start.real
    Is_year_end = date_val.is_year_end.real
    Is_year_start = date_val.is_year_start.real
    Is_leap_year = date_val.is_leap_year.real
    Year = date_val.year
    Month = date_val.month
    
    return Day, DayofWeek, Dayofyear, Week, Is_month_end, Is_month_start, Is_quarter_end, Is_quarter_start, Is_year_end, Is_year_start, Is_leap_year, Year, Month
```

```python
def app(data):
    funct = lambda x: pd.Series(extract_date_features(x))
    data[['Day', 'DayofWeek', 'DayofYear', 'Week', 'Is_month_end', 'Is_month_start',
      'Is_quarter_end', 'Is_quarter_start', 'Is_year_end', 'Is_year_start', 'Is_leap_year', 'Year', 'Month']] = data.Date_Col.apply(funct)
    return data
```

```python
app(hdfc)
```

```python
app(bhp)
```

```python
app(cipla)
```

```python
app(coromandel)
```

```python
app(tcs)
```
